```python
import os
from selenium.webdriver import Chrome
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
import time
import pandas as pd

driver = Chrome()

urls=[#list of urls for each company]

for url in urls:
    driver.get(url)
    pageno=1
    driver.delete_all_cookies()
    #driver.refresh()
    time.sleep(1)

    driver.find_element_by_xpath("//a[@class='susiLink sign-in']").click()
    time.sleep(1)
    driver.find_element_by_xpath("//input[@id='userEmail']").send_keys("username")
    time.sleep(1)
    driver.find_element_by_xpath("//input[@id='userPassword']").send_keys("password")
    time.sleep(1)
    driver.find_element_by_xpath("//button[@class='gd-ui-button minWidthBtn css-1sdotxz']").click()
    head = False
    time.sleep(5)
    while True:
        print(pageno)
        driver.get(url)
        time.sleep(5)
        divs = driver.find_elements_by_xpath("//ol[@class=' highlightsActive empReviews emp-reviews-feed pl-0']/li")
        print(len(divs))
        dataa = []
        for div in divs:
            val = {}
            try:
                val['title'] = div.find_element_by_xpath(".//a[@class='reviewLink']").text
            except:
                val['title'] = ""
            try:
                val['rating'] = div.find_element_by_xpath(".//span[@class='value-title']").get_attribute("title")
            except:
                val['rating'] = "" 
            try:
                aa=[]
                for element in div.find_elements_by_xpath(".//span[@class='gdBars gdRatings med']"):
                    aa.append(element.get_attribute("title"))
                val['work/life balance rating'] = aa
            except:
                val['work/life balance rating'] = ""
            try:
                div.find_element_by_xpath(".//div[@class='v2__EIReviewDetailsV2__continueReading v2__EIReviewDetailsV2__clickable']").click()
                comments = div.find_elements_by_xpath(".//div[@class='v2__EIReviewDetailsV2__fullWidth ']")
                review = []
                for comment in comments:
                    review.append(comment.get_attribute("innerText"))
                val['reviews'] = review        
            except:
                comments = div.find_elements_by_xpath(".//div[@class='v2__EIReviewDetailsV2__fullWidth ']")
                review = []
                for comment in comments:
                    review.append(comment.get_attribute("innerText"))
                val['reviews'] = review
            dataa.append(val)
        df = pd.DataFrame(dataa)
        if head:
            df.to_csv("Capegemini.csv", mode='a', index=False)
            head = False
        else:
            df.to_csv("Capegemini.csv", mode='a', index=False, header=False)    
            
        if (pageno>150):
            break

        driver.delete_all_cookies()
        url=driver.find_element_by_xpath("//a[@class='pagination__ArrowStyle__nextArrow  ']").get_attribute("href")
        print(url)
        pageno=pageno+1
        time.sleep(15)
        if url==None:
            break


    time.sleep(5)

    driver.quit()
```

```python
def f1(dff):
    try:
        a=dff[2]
    except:
        return None
    return int(a)
def f2(dff):
    try:
        a=dff[9]
    except:
        return None
    return int(a)
def f3(dff):
    try:
        a=dff[16]
    except:
        return None
    return int(a)
def f4(dff):
    try:
        a=dff[23]
    except:
        return None
    return int(a)
def f5(dff):
    try:
        a=dff[30]
    except:
        return None
    return int(a)

def reviews_re(text):
    import re
    try:
        text=text.replace("\"","'")
        text=text.lstrip("['Pros\\n\n")
        text=text.rstrip("']")
        text=re.split(r"\', 'Cons\\n\\n", text)
    except:
        return None
    return text

def pro(l):
    try:
        a=l[0]
    except:
        return None
    if a=='':
        return None
    return a
def con(l):
    try:
        a=l[1]
    except:
        return None
    return a    

def modified_reviews_csv(company):
    adobe1=pd.read_csv("{}.csv".format(company))
    categories=['work/life','culture/values','career_opp','compensation/ben','senior_mng']
    adobe1 = adobe1.reindex(columns = adobe1.columns.tolist() + ['work/life','culture/values','career_opp','compensation/ben','senior_mng','pros','cons'])
    adobe1['work/life']=adobe1['work/life balance rating'].apply(f1)
    adobe1['culture/values']=adobe1['work/life balance rating'].apply(f2)
    adobe1['career_opp']=adobe1['work/life balance rating'].apply(f3)
    adobe1['compensation/ben']=adobe1['work/life balance rating'].apply(f4)
    adobe1['senior_mng']=adobe1['work/life balance rating'].apply(f5)
    for i in categories:
        adobe1[i]=adobe1[i].fillna(adobe1['rating'])
    adobe1=adobe1.drop('work/life balance rating',axis=1)
    adobe1['reviews']=adobe1['reviews'].apply(reviews_re)
    adobe1['pros']=adobe1['reviews'].apply(pro)
    adobe1['cons']=adobe1['reviews'].apply(con)
    adobe1=adobe1.drop('reviews',axis=1)
    adobe1=adobe1.dropna(axis=0, subset=['pros', 'cons'], thresh=1)
    adobe1['title']=adobe1['title'].apply(lambda x: x.replace("\"",""))
    adobe1.to_csv('{}_modified.csv'.format(company), index=False)

    
import os
os.chdir(r'C:\Users\lenovo\Desktop')
import pandas as pd
companies=[#list]
for company in companies:
    modified_reviews_csv(company)
```

```python
for company in companies:
    df=pd.read_csv('{}_modified.csv'.format(company))
    df.to_csv("combined.csv", mode='a', index=False)
```
